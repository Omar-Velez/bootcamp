# <strong> Introducción a la IA y  Tipos de Aprendizaje </strong>

## <strong> 📜Historia y Evolución de la IA </strong>

La historia de la inteligencia artificial (IA) comenzó en la antigüedad con mitos sobre seres artificiales inteligentes, pero su desarrollo moderno se remonta a mediados del siglo XX. En la década de 1940, la invención de la computadora digital programable sentó las bases para la IA. En 1950, Alan Turing publicó su artículo "Computing Machinery and Intelligence", introduciendo el concepto del test de Turing, que evalúa la capacidad de una máquina para exhibir comportamiento inteligente indistinguible del de un ser humano.

En 1956, la conferencia en Dartmouth College marcó el nacimiento formal de la IA como disciplina académica, liderada por John McCarthy, Marvin Minsky, Nathaniel Rochester y Claude Shannon. La IA ha experimentado avances significativos, incluyendo el desarrollo del aprendizaje automático y el aprendizaje profundo, así como períodos de desilusión financiera conocidos como los "inviernos de la IA".

## <strong> 🫧 Diferencias entre IA, ML y Deep Learning </strong>

- **Inteligencia Artificial (IA):** Se refiere a la creación de sistemas que pueden realizar tareas que normalmente requieren inteligencia humana. La IA abarca un amplio rango de técnicas y disciplinas, desde el procesamiento del lenguaje natural hasta la visión artificial.

- **Machine Learning (ML):** Es un subconjunto de la IA que se enfoca en el desarrollo de algoritmos que permiten a las máquinas aprender de los datos sin ser programadas explícitamente. El ML se utiliza para predecir resultados o clasificar datos.

- **Deep Learning (DL):** Es una rama del ML que utiliza redes neuronales artificiales para analizar datos complejos, como imágenes y audio. El DL ha sido crucial en avances recientes en áreas como la visión artificial y el procesamiento del lenguaje natural.

## <strong> 🧑🏻‍🏫Tipos de Aprendizaje </strong>

- **Aprendizaje Supervisado**: El aprendizaje supervisado es un paradigma donde el modelo es entrenado con datos etiquetados. Se utilizan entradas y sus correspondientes salidas deseadas para aprender una función de mapeo.

    - Regresión Lineal para predecir valores clínicos en pacientes.

    - Clasificación con Naive Bayes para detectar enfermedades en imágenes médicas.

- **Aprendizaje No Supervisado**: En el aprendizaje no supervisado, los datos no están etiquetados. El modelo encuentra patrones ocultos o estructuras en los datos.

 - K-Means para segmentación de pacientes según características clínicas.

 - Análisis de Componentes Principales (PCA) para reducción de dimensionalidad en datos de laboratorio.

- **Aprendizaje Reforzado**: El aprendizaje reforzado se basa en la interacción de un agente con un entorno para maximizar una recompensa.

 - Algoritmos de IA para optimizar planes de tratamiento personalizados.

 - Control de dispositivos médicos autónomos.

# <strong> Historia y Ciclo de Vida de una Aplicación de Aprendizaje de Máquina </strong>

<center><img src="https://github.com/juliandariogiraldoocampo/ia_taltech/blob/main/fiscalia/assets/Ciclo_Vida_Aplicacion_ML.png?raw=true" width=65%> </center>

El Aprendizaje de Máquina **(ML)** ha evolucionado significativamente desde sus inicios, pasando de ser una disciplina académica a una herramienta fundamental en la industria. A lo largo de su historia, el desarrollo de aplicaciones de ML ha seguido un ciclo de vida bien definido que garantiza la eficacia y la eficiencia de los modelos. A continuación, se detalla cada etapa de este ciclo:

<font size = 5>1. Investigación y Planificación </font>

Aunque la mayoría de la literatura no lo considera, es útil incluir una fase inicial de investigación y planificación. Esto implica:

- Explorar el dominio,
- Identificar oportunidades potenciales.
- Definir los objetivos generales del proyecto.

<font size = 5>2. Identificación del Problema </font>

La primera etapa implica identificar un problema o una oportunidad que pueda ser abordada mediante el aprendizaje de máquina. Esto requiere una comprensión profunda del negocio o del dominio en el que se está trabajando. Algunas preguntas clave para esta etapa incluyen:

- ¿Qué problema estamos tratando de resolver?
- ¿Cuáles son los objetivos clave?
- ¿Qué datos están disponibles para abordar este problema?

<font size = 5>3. Recolección de Datos </font>

Una vez identificado el problema, la siguiente etapa es la recolección de datos relevantes. Esto puede incluir datos estructurados, no estructurados o semiestructurados, dependiendo del tipo de problema. La calidad y la cantidad de los datos son cruciales para el éxito del modelo.

- Fuentes de datos: Bases de datos, sensores, APIs, archivos CSV, etc.
- Consideraciones éticas: Privacidad, consentimiento, sesgo en los datos.

<font size = 5>4. Preparación de Datos </font>

La preparación de los datos es una de las etapas más importantes y consumidoras de tiempo en el ciclo de vida de una aplicación de ML. Incluye limpieza, transformación y selección de características relevantes.

- Limpieza de datos: Manejo de valores faltantes, eliminación de datos duplicados.
- Transformación de datos: Normalización, escalado, codificación de variables categóricas.
- Selección de características: Identificar las variables más relevantes para el modelo.

<font size = 5>5. Ingeniería de Modelos </font>

En esta etapa, se desarrollan y entrenan los modelos de aprendizaje de máquina. Esto puede incluir desde modelos simples como regresión lineal hasta modelos más complejos como redes neuronales profundas.

- Selección del algoritmo: Depende del tipo de problema (clasificación, regresión, clustering) y de los datos disponibles.
- Entrenamiento del modelo: Uso de conjuntos de entrenamiento y validación para ajustar los hiperparámetros.

<font size = 5>6. Evaluación del Modelo </font>

La evaluación es crucial para determinar si el modelo cumple con los objetivos establecidos. Se utilizan métricas específicas según el tipo de problema, como precisión, recall, F1-score para clasificación, o R-squared para regresión.

- Métricas de evaluación: Selección de las métricas adecuadas para el problema.
- Validación cruzada: Técnica para evaluar el rendimiento del modelo en diferentes conjuntos de datos.

<font size = 5>7. Despliegue </font>

Una vez que el modelo ha sido evaluado y ajustado, se despliega en un entorno de producción. Esto puede implicar integrarlo con aplicaciones existentes o crear una nueva interfaz para que los usuarios interactúen con el modelo.

- Contenerización: Uso de herramientas como Docker para asegurar la consistencia del entorno.
- APIs: Creación de interfaces de programación de aplicaciones para interactuar con el modelo.

<font size = 5>8. Mantenimiento y Actualización </font>

El mantenimiento es una etapa continua que asegura que el modelo siga siendo relevante y efectivo a lo largo del tiempo. Esto incluye monitorear el rendimiento del modelo, actualizarlo con nuevos datos y ajustar los parámetros según sea necesario.

- Monitoreo del rendimiento: Seguimiento continuo de las métricas de evaluación.
- Reentrenamiento: Actualización periódica del modelo con nuevos datos para mantener su precisión.

---
🔃Este ciclo de vida es iterativo, y cada etapa puede influir en las demás. Por ejemplo, problemas encontrados durante la evaluación pueden requerir regresar a la etapa de preparación de datos o ingeniería de modelos.

# 📚 ¿QUÉ ES UN DICCIONARIO EN PYTHON?

Un **diccionario** es una estructura de datos que permite almacenar información en pares de **clave: valor**.

Es como un diccionario de verdad:
- Hay una **palabra** (la clave)
- Y el **significado** (el valor)

En programación funciona igual: Se busca algo por el nombre (clave) y se obtiene el dato asociado (valor).

## 🧳 Analogía aeroportuaria

Pensemos en un **panel de información de vuelos**:

| Clave (código de vuelo) | Valor (información asociada) |
|-------------------------|-------------------------------|
| AV123                   | "Bogotá - 10:30 AM - Puerta 5" |
| LA456                   | "Medellín - 11:15 AM - Puerta 3" |
| VH789                   | "Cali - 12:00 PM - Puerta 7" |

Así funciona un diccionario: **se almacena información asociada al código del vuelo como por ejemplo, la ciudad destino, la hora y la puerta de embarque**.

## 📋 Características principales:

- **Claves únicas:** No puede haber dos claves iguales (como no hay dos vuelos con el mismo código)
- **Acceso directo:** Se obtiene el valor rápidamente si se conoce la clave
- **Flexible:** Los valores pueden ser números, textos, listas, ¡incluso otros diccionarios!

## 🆚 Diferencia con las listas

| Listas | Diccionarios |
|--------|--------------|
| Los elementos se acceden por posición (índice 0, 1, 2...) | Los elementos se acceden por nombre (clave) |
| Se guardan datos en orden específico | Se guardan datos sin orden determinado |
| `lista[0]` → primer elemento | `diccionario["AV123"]` → valor asociado a esa clave |

## ✨ En resumen:

Un diccionario es como un **casillero con etiquetas**: cada casilla tiene un nombre (clave) y dentro se guarda lo que se necesita (valor). En el aeropuerto, cada código de vuelo (clave) tiene asociada toda su información operativa (valor).

## ✈️9. Diccionario de vuelos registrados

**Contexto:**

El área de operaciones del aeropuerto necesita almacenar y consultar información básica de varios vuelos programados durante la jornada.

**Problema:**

Crea un diccionario por vuelo con los campos: identificador, tipo de aeronave y estado operativo.  
Guarda 3 vuelos en una lista.  
Permite buscar un vuelo por su identificador e imprime su información.

**Solución esperada:**

Listas de diccionarios, búsqueda lineal.

**Retos:**

- Si el identificador no existe, imprime: “Vuelo no encontrado”.
- Trata de resolverlo usando:
  - una lista de diccionarios, o  
  - un diccionario de diccionarios.
- Considera que los identificadores pueden estar escritos con variaciones de mayúsculas y minúsculas (normalización con `.lower()`).


````python
# EJEMPLO DE SOLUCIÓN - 1
# Creamos una LISTA que contiene diccionarios, uno por cada vuelo

vuelos = [
    {"identificador": "HK4521",
     "tipo_aeronave": "Turbohélice",
     "estado": "En plataforma"},

    {"identificador": "HK3890",
     "tipo_aeronave": "Jet regional",
     "estado": "En aproximación"},

    {"identificador": "HK1203",
     "tipo_aeronave": "Aviación general",
     "estado": "Programado"}
]

# Solicitamos al usuario el identificador del vuelo a buscar
vuelo_buscado = input("Introduce el identificador del vuelo que deseas buscar: ")

# Búsqueda lineal del vuelo por identificador (normalizando mayúsculas/minúsculas)
vuelo_encontrado = None
for vuelo in vuelos:
    if vuelo["identificador"].lower() == vuelo_buscado.lower():
        vuelo_encontrado = vuelo
        break


# Mostramos la información del vuelo o un mensaje si no se encuentra
if vuelo_encontrado:
    print(f"\nInformación del vuelo:")
    print(f"Identificador: {vuelo_encontrado['identificador']}")
    print(f"Tipo de aeronave: {vuelo_encontrado['tipo_aeronave']}")
    print(f"Estado operativo: {vuelo_encontrado['estado']}")
else:
    print("Vuelo no encontrado.")
    print(vuelo_encontrado)
````

```csv
Vuelo, Tipo Aeronave, Pasajeros, Estado
```

**Reto adicional**

Modifica el código para que el usuario pueda definir **dinámicamente** el límite de pasajeros que activará la alerta, usando `input()`.

2. **Estadísticas descriptivas**
Calcula y muestra:
- **Total de vuelos** en el dataframe
- **Cantidad de vuelos con ALERTA** (superan el límite)
- **Porcentaje** que representan los vuelos con alerta sobre el total


**Conceptos nuevos**

| Concepto | Implementación |
|----------|----------------|
| **Filtrado booleano** | `df['Pasajeros'] > limite` |
| **Suma de booleanos** | `.sum()` cuenta los `True` |
| **Asignación condicional** | `df.loc[condición, columna] = valor` |


---

**Explicación detallada:**

| Concepto | ¿Qué hace? | Ejemplo |
|----------|------------|---------|
| **Filtrado booleano** | Crea una máscara de `True/False` según una condición | `df['Pasajeros'] > 60` → `[False, True, False, ...]` |
| **Suma de booleanos** | Cuenta los `True` (que valen 1) | `(df['Pasajeros'] > 80).sum()` → `33` |
| **Asignación condicional** | Modifica valores solo donde se cumple la condición | `df.loc[df['Pasajeros'] > 60, 'Informe Alerta'] = 'ALERTA'` |
| **Formato de strings** | Inserta variables en texto con formato controlado | `f'{porcentaje:.1f}%'` → `"33.0%"` |

````python
import pandas as pd

# CARGA, CAPTURA E INICIALIZACIÓN DE DATOS
ruta = '/content/drive/MyDrive/data_IA_EOH/registro_vuelos.csv'
## skipinitialspace=True es para "saltar" los espacios iniciales en los nombres de las columnas
df = pd.read_csv(ruta, skipinitialspace=True)
pax_limit = 60

# ANÁLISIS Y PROCESAMIENTO
# Crear nueva columna para evitar que en la condición se generen datos NaN
df['Informe Alerta'] = ''

# Filtrar los registros en los que "Pasajeros" es superior al pax_limit y asignar 'ALERTA' en la nueva columna
df.loc[df['Pasajeros'] > pax_limit, 'Informe Alerta'] = 'ALERTA'

# RETO
# Contar el número de veces que se cumple la condición
total_registros = len(df)
total_alertas = (df['Pasajeros'] > pax_limit).sum()
tasa_alertas = 100*(total_alertas / total_registros)

# VISUALIZACIÓN
display(df.tail(5))
print(f'''
REPORTE __________________________________________
📒Total de registros:     {total_registros}
🚨Número de Alertas:      {total_alertas}
⚠️Tasa de Incidentes:     {tasa_alertas:.2f}%
___________________________________________________
🛬 PAX_LIMIT:             {pax_limit}
''')
````

La nueva columna debe contener:

- **Vacío** (o `""`) cuando no se cumpla ninguna de las condiciones



**Solución esperada:**

Utilizar filtrado condicional con pandas para crear la columna y asignar los valores correspondientes.

```python
# Inicializar la columna 'Riesgo Operativo' vacía
df['Riesgo Operativo'] = ''

# Asignar 'Sí' a los vuelos que cumplen ambas condiciones
df.loc[(df['Pasajeros'] > 60) & (df['Estado'] == 'En aproximación'), 'Riesgo Operativo'] = 'Sí'
```

````python
import pandas as pd

# CARGA, CAPTURA E INICIALIZACIÓN DE DATOS
ruta = '/content/drive/MyDrive/data_IA_EOH/registro_vuelos.csv'
## skipinitialspace=True es para "saltar" los espacios iniciales en los nombres de las columnas
df = pd.read_csv(ruta, skipinitialspace=True)
pax_limit = 60

# ANÁLISIS Y PROCESAMIENTO
# Crear nueva columna para evitar que en la condición se generen datos NaN
df['Riesgo Operativo'] = ''

# Filtrar los registros que cumplan ambas condiciones
cond_1 = df['Pasajeros'] > pax_limit
cond_2 = df['Estado'] == 'En aproximación'

df.loc[cond_1 & cond_2, 'Riesgo Operativo'] = 'SI'

# RETO
# Contar el número de veces que se cumple la condición
total_registros = len(df)
total_alertas = (df['Riesgo Operativo'] == 'SI').sum()
tasa_alertas = 100*(total_alertas / total_registros)

# VISUALIZACIÓN
display(df.tail(5))
print(f'''
REPORTE __________________________________________
📒Total de registros:       {total_registros}
🚨Número de Alertas:        {total_alertas}
⚠️Tasa de Riesgo Operativo: {tasa_alertas:.2f}%
___________________________________________________
🛬 PAX_LIMIT:               {pax_limit}
''')

df.to_excel('Informe Riesgo Operativo.xlsx', index=False)
````

<font size = 4>**Columnas del archivo .csv**</font>

| **Variable**                 | **Tipo de dato** | **Descripción**                                               |

| `Retraso_Minutos`          | entero           | Retraso acumulado en minutos                                |

<br>

<font size = 4>**Problema**</font>

Se desea calcular una nueva medida llamada **Tasa de Presión Operativa**, definida como el:

<font color = #0BE5CF>**cociente entre el número de pasajeros y las horas transcurridas desde la programación del vuelo.**</font>

Es decir:

```python
Tasa_Presión_Operativa = Pasajeros / Horas_Desde_Programación
```

Sin embargo, existen vuelos que acaban de ser programados (`Horas_Desde_Programación = 0`).  
En estos casos, la tasa se definirá como igual al número de pasajeros, para evitar divisiones por cero.

<br>

**Retos**

- Calcular el promedio general de la tasa de presión operativa en toda la muestra.
- Crear un nuevo dataset con los vuelos que presentan tasas altas y podrían considerarse en riesgo de congestión operativa.


````python
import pandas as pd
import matplotlib.pyplot as plt

# CARGA Y CAPTURA DE DATOS
ruta = 'https://github.com/juliandariogiraldoocampo/ia_taltech/raw/refs/heads/main/aeropuerto/vuelos_operacion_datos_sinteticos.csv'

df = pd.read_csv(ruta)

# ANÁLISIS Y PROCESAMIENTO
# Cálculo de Nuevos Campos/Columnas
df['Tasa Presion Operativa'] = df['Pasajeros']
condicion = df['Horas_Desde_Programación'] > 0
df.loc[condicion, 'Tasa Presion Operativa'] = df['Pasajeros'] / df['Horas_Desde_Programación']

# Cantidad de vuelos que tienen una Tasa de Presión Operativa mayor a 30
# total_alertas = (df['Riesgo Operativo'] == 'SI').sum()
cant_tasa_operativa_alta = (df['Tasa Presion Operativa'] > 30).sum()

# Promedio Tasa de Presión Operativa
promedio_tasa = df['Tasa Presion Operativa'].mean()

# Crear nuevo DataFrame
lista_campos = ['ID_Vuelo', 'Pasajeros', 'Horas_Desde_Programación', 'Tasa Presion Operativa']
df_informe = df[lista_campos]
display(df_informe.head(5))
# VISUALIZACIÓN
print(f'''
____________________________ INFORME PRELIMINAR ____________________________

Cantidad de Vuelos con una Tasa Operativa mayor a 30:   {cant_tasa_operativa_alta}
Promedio General de la Tasa de Presión Operativa:       {promedio_tasa:.2f}
''')

plt.plot(df['Velocidad_Viento'])
plt.xlabel('Vuelo')
plt.ylabel('Tasa de Presión Operativa')
display(df.head(5))
````

# 📊 ¿Qué es el Análisis Exploratorio de Datos (EDA)?

El Análisis Exploratorio de Datos es una etapa inicial del trabajo con información cuyo propósito es **comprender los datos antes de tomar decisiones o construir modelos**. En esta fase se examinan valores, distribuciones, tendencias, posibles errores y relaciones entre variables.

El EDA permite responder preguntas como:

- ¿Qué tan dispersos están los datos?
- ¿Existen valores atípicos o inconsistencias?
- ¿Qué patrones comienzan a observarse?
- ¿Hay relaciones evidentes entre variables?

Para ello se utilizan herramientas estadísticas básicas (promedios, máximos, mínimos, conteos) y visualizaciones que facilitan la interpretación.

Más que producir conclusiones definitivas, el análisis exploratorio busca **entender el comportamiento general de la información**, identificar posibles problemas en los datos y orientar los siguientes pasos del análisis.

En síntesis, es el proceso de “conversar con los datos” antes de analizarlos de manera más avanzada.
